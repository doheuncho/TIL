<h1>1.유효한 팰린드롬</h1>
<p>리트코드 125번</p>
[링크](https://leetcode.com/problems/valid-palindrome/)

```python
# 1.리스트를 이용한 풀이
# Runtime : 284 ms  Memory : 19.6 MB

class Solution:
    def isPalindrome(self, s: str) -> bool:
        s_list = []
        for char in s:
            if char.isalnum():
                s_list.append(char.lower())

        while len(s_list) > 1:
            if s_list.pop(0) != s_list.pop():
                return False

        return True
```

```python
# 2.데크를 이용한 풀이
# leetcode는 아래 모듈들을 미리 import함
# Runtime : 84 ms  Memory : 19.2 MB
import collections
from typing import *


class Solution:
    def isPalindrome(self, s: str) -> bool:
        s_deque: Deque = collections.deque()

        for char in s:
            if char.isalnum():
                s_deque.append(char.lower())

        while len(s_deque) > 1:
            if s_deque.popleft() != s_deque.pop():
                return False

        return True
```

```python
# 3.문자열 슬라이싱을 이용한 풀이
# leetcode는 아래 모듈들을 미리 import함
# Runtime : 40 ms  Memory : 15.7 MB
import re


class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        # a-z, 0-9 를 제외한(^)문자들을 ''로 치환
        s = re.sub('[^a-z0-9]', '', s)

        # s를 뒤집은 문자열이 s와 같은지 판별
        return s == s[::-1]
```

<h1>2.문자열 뒤집기</h1>
<p>리트코드 344번</p>
[링크](https://leetcode.com/problems/reverse-string/)

```python
# 1. 투 포인터를 이용한 풀이
# Runtime : 298 ms  Memory : 18.3 MB
from typing import *


class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        leftpointer, rightpointer = 0, len(s) - 1
        
        while leftpointer < rightpointer:
            s[leftpointer], s[rightpointer] = s[rightpointer], s[leftpointer]
            leftpointer += 1
            rightpointer -= 1
```

```python
# 2. 파이썬다운 방식(Pythonic way)
# Runtime : 201 ms  Memory : 18.4 MB

class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        s.reverse()
```

<h1>3.로그 파일 재정렬</h1>
<p>리트코드 937번</p>
[링크](https://leetcode.com/problems/reorder-data-in-log-files/)

```python
# 람다와 +연산자를 활용한 풀이
# Runtime : 41 ms  Memory : 14.4 MB

class Solution:
    def reorderLogFiles(self, logs: List[str]) -> List[str]:
        letters, digits = [], []
        for log in logs:
            if log.split()[1].isdigit():
                digits.append(log)
            else:
                letters.append(log)
        
        letters.sort(key = lambda x: (x.split()[1:], x.split()[0]))
        return letters + digits
```

<h1>4.가장 흔한 단어</h1>
<p>리트코드 819번</p>
[링크](https://leetcode.com/problems/most-common-word/)

```python
# 1.리스트 컴프리헨션, counter 객체를 사용한 풀이
# leetcode는 아래 모듈들을 미리 import함
# Runtime : 47 ms  Memory : 14.5 MB
import re
from typing import *
import collections


class Solution:
    def mostCommonWord(self, paragraph: str, banned: List[str]) -> str:
        # paragraph안의 단어(\w)를 제외(v)한 성분을 공백(' ')으로 치환하고 나눈뒤, banned에 없는 원소들을 나눈 리스트 생성
        words = [word for word in re.sub(r'[^\w]', ' ', paragraph)
                 .lower().split()
                 if word not in banned]

        # Counter(list) : 리스트 내의 원소를 세어 dic형태로 반환
        counts = collections.Counter(words)

        return counts.most_common(1)[0][0]
```

<h1>5.그룹 애너그램</h1>
<p>리트코드 49번</p>
[링크](https://leetcode.com/problems/group-anagrams/)

```python
#1.정렬 후 딕셔너리를 이용한 풀이
# leetcode는 아래 모듈들을 미리 import함
# Runtime : 158 ms  Memory : 17.1 MB
from typing import *
import collections

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        # defaltdic() : ()안의 값을 defalt로 한 딕셔너리를 생성. 이는 기존에 없는 key를 삽입할 때 keyError가 나는 것을 방지한다.
        anagrams = collections.defaultdict(list)    # 빈 리스트를 defalt로 한 딕셔너리 생성(새 key를 추가하면 {key : []}로 생성)

        for word in strs:
            anagrams[''.join(sorted(word))].append(word)    # word를 sort한 결과를 key로 가지는 value인 word를 추가해줌
        
        return list(anagrams.values())
```

<h1>6.가장 긴 팰린드롬 부분 문자열</h1>
<p>리트코드 5번</p>
[링크](https://leetcode.com/problems/longest-palindromic-substring/)

```python
# 투 포인터를 활용한 풀이
# Runtime : 287 ms  Memory : 14.2 MB

class Solution:
    def longestPalindrome(self, s: str) -> str:
        # 예외 처리
        if len(s) < 2 or s == s[::-1]:
            return s
        
        # 투 포인터 확장
        def expand(left: int, right: int) -> str:
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right +=1
            return s[left + 1:right]
        
        result = ''
        for i in range(len(s) - 1):
            result = max(result,
                         expand(i, i+1),
                         expand(i, i+2),
                         key = len)
        
        return result
```
