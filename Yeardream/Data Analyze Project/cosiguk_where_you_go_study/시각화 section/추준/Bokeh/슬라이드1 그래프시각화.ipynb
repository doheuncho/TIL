```python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure, show
output_notebook()
```

```python
df = pd.read_parquet("./main.parquet")
```

```python
print(df)
```

```python
count = np.array(df['일별이용자수'])
count
```

```python
dates = np.array(df['사용일자'], dtype=np.datetime64)
dates
```

```python
from bokeh.models import NumeralTickFormatter
from bokeh.models import HoverTool

p = figure(x_axis_type="datetime", title="일별이용자수", plot_height=300, plot_width=1000, tooltips="사용일자:datetime")
p.xgrid.grid_line_color=None
p.ygrid.grid_line_alpha=0.5
p.xaxis.axis_label = '사용일자'
p.yaxis.axis_label = '일별이용자수(만)'
p.yaxis.formatter = NumeralTickFormatter(format="0")
p.line(dates, df['일별이용자수']// 10000,color="#666600")
show(p)
```
