```python
import numpy as np
import pandas as pd
```

```python
aa = pd.read_parquet("main.parquet")
```

```python
aa = aa.sort_values("사용일자").reset_index(drop=True)
```

```python
aa
```

```python
aa["요일"] = np.nan
```

```python
for i in range(len(aa["요일"])):
    if i % 7 == 0 or i % 7 == 6 :
        aa["요일"][i] = "주말"
    else:
        aa["요일"][i] = "평일"
aa
```

```python
aa = aa.drop(0)
```

```python

aa.tail(10)
```

```python
data = {'주간시작일자': [],
        '주말평균승차': [],
        '평일평균승차': []}

new_data = pd.DataFrame(data)


new_data
```

```python
sum = 0
count = 0

for i in range(len(aa["요일"])):
    if i % 7 == 0:
        new_data = new_data.append({'주간시작일자':aa["사용일자"][i]},ignore_index=True)
        sum += aa["일별이용자수"][i]
    elif 0<(i % 7)<4:
        sum += aa["일별이용자수"][i]
    elif i % 7 == 4:
        sum += aa["일별이용자수"][i]
        new_data["평일평균승차"][count] = sum/5
        sum = 0
    elif i % 7 == 5:
        sum += aa["일별이용자수"][i]
    else:
        sum += aa["일별이용자수"][i]
        new_data["주말평균승차"][count] = sum/2
        count += 1

new_data
```

```python
new_data["주말평균승차"][199] = 3730136
```

```python
new_data["평일-주말"] = new_data["평일평균승차"]- new_data["주말평균승차"]
new_data
```

```python
new_data.to_csv("평일-주말승차.csv")
```
