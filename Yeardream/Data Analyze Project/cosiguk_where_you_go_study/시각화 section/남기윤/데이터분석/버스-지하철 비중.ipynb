```python
import numpy as np
import pandas as pd
import datetime as datetime
```

```python
bus = pd.read_parquet("bus_second.parquet/bus_per_week.parquet")
sub = pd.read_parquet("main.parquet/main.parquet")
bus
```

```python
subs = sub.sort_values("사용일자").reset_index(drop=True)
subs
```

```python
data = {'주간시작일자': [],
        '주별이용자수': []}

new_data = pd.DataFrame(data)

new_data
```

```python
sum = 0
count = 0
for i in range(len(subs["사용일자"])):
    if i % 7 == 0:
        new_data = new_data.append({'주간시작일자':subs["사용일자"][i]},ignore_index=True)
        sum += subs["일별이용자수"][i]
    elif 0<(i % 7)<6:
        sum += subs["일별이용자수"][i]
    elif i % 7 == 6:
        sum += subs["일별이용자수"][i]
        new_data["주별이용자수"][count] = sum
        sum = 0
        count +=1
new_data.rename(columns = {'주별이용자수' : '지하철'}, inplace = True)
new_data
```

```python
bus_sub = pd.concat([bus,new_data['지하철']],axis=1)
bus_sub
```

```python
bus_sub.rename(columns = {'주별이용자수' : '버스'}, inplace = True)
bus_sub
```

```python
bus_sub.to_csv("버스-지하철 비중.csv")
```

```python
bus_sub["버스비율"] = np.nan
bus_sub["지하철비율"] = np.nan
bus_sub["버스비율"] = bus_sub["버스"]/(bus_sub["버스"]+bus_sub["지하철"])
bus_sub["지하철비율"] = bus_sub["지하철"]/(bus_sub["버스"]+bus_sub["지하철"])
bus_sub
bus_sub.to_csv("버스-지하철 비중.csv")
```
