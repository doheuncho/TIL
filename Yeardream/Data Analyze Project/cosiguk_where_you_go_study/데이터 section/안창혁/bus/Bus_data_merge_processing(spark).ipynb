# spark의 빠름을 알아보자!

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
import pandas as pd
import numpy as np
import glob
import os
```

### 2021년

```python
#csv 합치기
input_file = r"D:data/2021/"
output_file = r'D:data/2021_bus_per_date.txt'

allFile_list = glob.glob(os.path.join(input_file,'BUS_STATION_BOARDING_MONTH_*'))
print(allFile_list)
allData = []
for file in allFile_list:
    df = pd.read_csv(file, encoding="euc-kr",names = ["사용일자","노선번호","노선명","버스정류장ARS번호","역명","승차총승객수","하차총승객수","등록일자"], skiprows=[0])
    allData.append(df)

dataCombine = pd.concat(allData, axis=0, ignore_index=True)
dataCombine.to_csv(output_file,encoding='euc-kr', index=False)
```

```python
pq = spark.read.csv('./data/2021_bus_per_date.txt', encoding="euc-kr", header=True)
pq.show()
```

```python
#csv 확인
pq.select("사용일자").distinct().collect()
```

```python
#parquet 저장
pq.write.parquet('2021_bus_per_date')
```
