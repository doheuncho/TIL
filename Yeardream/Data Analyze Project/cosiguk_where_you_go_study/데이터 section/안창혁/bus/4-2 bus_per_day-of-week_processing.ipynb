## 1. csv merge

```python
import pandas as pd
import numpy as np
```

```python
#csv 합치기
import glob
import os

input_file = r'./data4'
output_file = r'./data4/all_bus_per_day-of-week.csv'

allFile_list = glob.glob(os.path.join(input_file, 'bus_*'))
print(allFile_list)
allData = []
for file in allFile_list:
    df = pd.read_csv(file)
    allData.append(df)

dataCombine = pd.concat(allData, axis=0, ignore_index=True)
dataCombine.to_csv(output_file, index=False)
```

```python
#합친 csv 불러오기
df = pd.read_csv("./data4/all_bus_per_day-of-week.csv")
df
```

## 2. 데이터 전처리

```python
#데이터 타입 확인
df.dtypes
```

```python
#필요없는 컬럼 제거
df = df.drop(["항목","전월하루평균(명)", "전월대비(명) ±"], axis=1)
df
```

```python
#필요없는 컬럼값 제거
df = df[df.요일 != "하루 평균"].reset_index(drop=True)
df
```

```python
#컬럼명 이름 변경
df = df.rename(columns={"하루평균(명)":"평균승차인원"})
df
```

```python
#날짜 데이트 타입변형
#이유 : int타입이므로 그룹화했을 때 날짜도 평균으로 나타남. 그러므로 string으로 바꿔줌.
df = df.astype({"날짜" : str})
df.dtypes
```

```python
#지역과 요일별로 그룹화하기
df2 = df.groupby(["지역","요일"]).mean().round(0).reset_index()
df2
```

```python
#요일 정렬하기
#요일을 정렬할 때 string이므로 알파벳순으로 정렬이 된다.
#따라서 요일별로 정렬할 수 있도록 모듈을 불러온다.
from pandas.api.types import CategoricalDtype
```

```python
#요일을 월요일 순서대로 정렬할도록 설정
cat_day_of_week = CategoricalDtype(
    ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'], 
    ordered=True
)
#요일 타입이 카테고리 타입으로 변경되었는지 확인
df2["요일"] = df2["요일"].astype(cat_day_of_week)
df2.dtypes
```

```python
#지역과 요일기준으로 정렬
df2 = df2.sort_values(["지역", "요일"]).reset_index(drop=True)
df2
```

```python
#parqeut 저장
df2.to_parquet("bus_per_day-of-week.parquet")
```
