# Cosiguk project - AWS 에서 Flask 배포ㅇㅅㅇ

* 사실 AWS EC2 들어가서 flask 실행하는게 전부임.  

  
* aws 로그인 > ec2 > 인스턴스 시작  
  
* ubuntu 18.04  선택. 왜 이거 쓰는지 모름. 아무튼 이거.
  
* 돈쓰기 싫으니까 프리티어 선택 ㄱㄱ  
  
* 모두 기본값으로 설정하고 보안 그룹만 편집하자.  
  
  
* 포트 범위 5000, 
  
* 새 키페어 선택. 이름은 알아서 적절하게 짓는다.
* 키페어 다운로드 위치는 정확하게 알고 있어야 한다. 인스턴스 연결 할때 필요.
* 인스턴스 시작을 누르고 실행하는데 5분 정도 걸림.  
  
* 뭘 검사하는진 모르지만 2개 검사 완료되면 드가자.

* 해당 인스턴스 클릭 > 연결 ㄱ
* cmd를 관리자로 실행. 키페어가 위치한 경로로 이동한다. 
  

  
* ssh 주소 복붙.  

* 그리고 아래 내용들을 설치, 실행 == 끝
* 중간중간 yes 또는 y 입력 해줘야됨.



sudo apt update  
sudo apt install python3-pip  
pip3 install flask

git clone https://yeardream-gitlab.elice.io/yeardream-project/project-10/cosiguk-where-you-go.git  
git 이름 입력  
git password 입력

cd cosiguk-where-you-go/Cosiguk  
sudo python3 app.py

  
* 저기 cmd에 적힌 주소는 내부에서 접속하는 주소이므로 외부에서는 아무리 주소창에 입력해도 들어갈 수 없다.  

* 13.124.37.47:5000 << 이 주소 + 포트를 주소창에 입력해야 함  

* 결과창 ㅇㅇ
* 프롬프트 창을 닫고도 계속 실행되게 하고 싶다면 sudo python3 app.py 대신에 다음코드를 입력하자.  
    nohup python3 -u app.py &  ( & 까지 입력 해줘야함.)
