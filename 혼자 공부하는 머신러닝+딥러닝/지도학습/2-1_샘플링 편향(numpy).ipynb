```python
import numpy as np      #샘플링을 위해 nupy를 import

input_arr = np.array(Fish_data)     #파이썬 리스트를 array()함수를 통해 numpy배열로 변경
target_arr = np.array(Fish_target)

print(input_arr)
```

```python
print(input_arr.shape)      #.shape는 (샘플 수, 특성  수)를 출력
```

```python
index = np.arange(49)     #0~48까지의 인덱스를 생성
np.random.seed(42)      #랜덤시드값을 정해두면 실행할때마다 같은값을 얻음
np.random.shuffle(index)      #생성한 인덱스를 섞음

Train_input = input_arr[index[:35]]
Train_target = target_arr[index[:35]]

Test_input = input_arr[index[35:]]
Test_target = target_arr[index[35:]]
```

```python
import matplotlib.pyplot as plt     #그래프를 통해 산점도를 확인

plt.scatter(Train_input[:, 0], Train_input[:, 1])     #Train_input 전체([:])에서 0, 1번 특성을 추출
plt.scatter(Test_input[:, 0], Test_input[:, 1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn = KNeighborsClassifier()
kn = kn.fit(Train_input, Train_target)      #numpy로 샘플링한 데이터를 학습시킴

kn.score(Test_input, Test_target)     #나머지 데이터로 정확도 측정. 샘플링 편향이 없어져 문제 해결
```
